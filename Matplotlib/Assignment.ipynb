```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-bright")
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
Virat_kohali = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
Virendra_Sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
Rohit_Sharma = [500, 600, 0, 1196, 1293, 815, 564, 1293, 1030, 1490]
plt.plot(years, Virat_kohali, color='pink', marker='o', linestyle=':', label= "Kohali") 
plt.plot(years, Virendra_Sehwag, color='yellow', marker='8', linestyle='-.', label= "Sehwag")
plt.plot(years, Rohit_Sharma, color='green', marker='p', linestyle='--', label= "Rohit")
plt.title("Epic Battle of the Batsmen")
plt.grid(True)
plt.tight_layout()
plt.legend()
```

```python
import numpy as np
years = [2010, 2011, 2012, 2013, 2014]
kohali = [0, 0, 500, 800, 1100]
sehwag = [0, 300, 800, 1200, 1500]
sachin = [500, 700, 1100, 1500, 1800]
x = np.arange(len(years))
width = 0.25
```

```python
plt.bar(x-width, kohali, width=width, label="Kohali")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x+width, sachin, width=width, label="Sachin")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
```
